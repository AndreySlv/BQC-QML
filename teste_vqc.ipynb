```python
#ESSE É O CERTO - FAKE BACKEND 
import time
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from qiskit.primitives import Sampler
from qiskit_ibm_runtime.fake_provider import FakeBrisbane
from qiskit_ibm_runtime import QiskitRuntimeService, Session
from qiskit.circuit.library import EfficientSU2, ZZFeatureMap
from qiskit_machine_learning.algorithms.classifiers import VQC
from qiskit_machine_learning.optimizers import COBYLA
from qiskit_machine_learning.utils import algorithm_globals

# ====================== Dados ======================
iris = load_iris()
X = MinMaxScaler().fit_transform(iris.data)
y = iris.target


# Dividir treino/teste
algorithm_globals.random_seed = 128
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=128)

# ====================== Circuitos ======================
num_qubits = X.shape[1]
feature_map = ZZFeatureMap(feature_dimension=num_qubits, reps=1)
ansatz = EfficientSU2(num_qubits, reps=1)

# ====================== Backend e Session ======================
backend = FakeBrisbane()
print(f"Usando backend: {backend.name}")

with Session(backend=backend) as session:
    sampler = Sampler()
    
    # Callback
    objective_vals = []
    def callback(weights, obj_value):
        clear_output(wait=True)
        objective_vals.append(obj_value)
        plt.plot(objective_vals)
        plt.title("Objective Function")
        plt.xlabel("Iteration")
        plt.ylabel("Loss")
        plt.show()
        
    # ====================== VQC ======================
    optimizer = COBYLA(maxiter=100)
    vqc = VQC(
        sampler=sampler,
        feature_map=feature_map,
        ansatz=ansatz,
        optimizer=optimizer,
        callback=callback,
    )

    print("Treinando VQC...")
    start = time.time()
    vqc.fit(X_train, y_train)
    duration = round(time.time() - start)
    print(f"Treino finalizado em {duration} segundos")
        
    
# ====================== Avaliação ======================
accuracy_train = vqc.score(X_train,y_train)
accuracy_test = vqc.score(X_test, y_test)
print(f"Acurácia no treinamento: {accuracy_train}")
print(f"Acurácia no teste: {accuracy_test}")
```
